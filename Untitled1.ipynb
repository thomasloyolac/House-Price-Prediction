```python
x=df1
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=100)
lr=LinearRegression()
lr.fit(xtrain,ytrain)
print(lr.score(xtrain,ytrain))
print(lr.score(xtest,ytest))
ypred=lr.predict(xtest)
print(np.sqrt(mean_squared_error(ytest,ypred)))
```

```python
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
```

```python
rf=RandomForestRegressor()
rf.fit(xtrain,ytrain)
print(rf.score(xtrain,ytrain))
print(rf.score(xtest,ytest))
ypred=rf.predict(xtest)
print(np.sqrt(mean_squared_error(ytest,ypred)))
```

```python
ad=AdaBoostRegressor()
ad.fit(xtrain,ytrain)
print(ad.score(xtrain,ytrain))
print(ad.score(xtest,ytest))
ypred=ad.predict(xtest)
print(np.sqrt(mean_squared_error(ytest,ypred)))
```

```python
gb=GradientBoostingRegressor()
gb.fit(xtrain,ytrain)
print(gb.score(xtrain,ytrain))
print(gb.score(xtest,ytest))
ypred=gb.predict(xtest)
print(np.sqrt(mean_squared_error(ytest,ypred)))
```

```python
from sklearn.tree import DecisionTreeRegressor
dd=DecisionTreeRegressor()
dd.fit(xtrain,ytrain)
print(dd.score(xtrain,ytrain))
print(dd.score(xtest,ytest))
ypred=dd.predict(xtest)
print(np.sqrt(mean_squared_error(ytest,ypred)))
```

```python
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor()
kn.fit(xtrain,ytrain)
print(kn.score(xtrain,ytrain))
print(kn.score(xtest,ytest))
ypred=kn.predict(xtest)
print(np.sqrt(mean_squared_error(ytest,ypred)))
```
